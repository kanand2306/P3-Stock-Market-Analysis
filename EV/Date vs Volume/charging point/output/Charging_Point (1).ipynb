```python
from kafka import KafkaConsumer
import matplotlib.pyplot as plt
import json
import threading
```

```python
consumer = KafkaConsumer('powergrid1',
                         group_id='powergrid1',
                         bootstrap_servers=['localhost:9092'],
                        )
print("consumer started ...")
```

```python
x = {}
def plot():
    global x
    for message in consumer:
        x[json.loads((message.value).decode("utf-8"))["date"]] = float(json.loads((message.value).decode("utf-8"))["volume"])
print(x)
```

```python
plot_thread = threading.Thread(target=plot)
plot_thread.start()
```

```python
fig = plt.figure(figsize = (12, 6))
#x = dict(sorted(x.items(), key=lambda item: item[0], reverse=False))
plt.plot([*x.keys()][:30], [*x.values()][:30])
plt.xlabel("Date", fontsize=15)
plt.ylabel("Volume", fontsize=15)
plt.xticks(rotation=90)
plt.title("POWERGRIDE SHARES", fontsize=15)
plt.show()
```
